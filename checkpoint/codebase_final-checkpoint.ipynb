## Visualising Data

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
).generate(str(data))
    fig = plt.figure(1, figsize=(15, 15))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
```

## Preprocessing Data

```python
from nltk.tokenize import TweetTokenizer
from sklearn.feature_extraction.text import TfidfVectorizer

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,GRU,LSTM,Embedding
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import SpatialDropout1D,Dropout,Bidirectional,Conv1D,Conv2D,GlobalMaxPooling1D,MaxPooling1D,MaxPool2D,Flatten
from tensorflow.keras.callbacks import ModelCheckpoint, TensorBoard, Callback, EarlyStopping
from tensorflow.keras.utils import to_categorical
from keras.regularizers import l2
```

```python
max_features = 5000
max_words = 22 # this could even be taken as 30 to include all words
num_classes=3
```

```python
train['text'].head()
```

```python
class nostemmimg():
    def stem(self,a):
        return a

from keras.preprocessing.text import text_to_word_sequence
# tokenization
# lementing
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# stemming
from nltk.stem import PorterStemmer
word_stemmer = PorterStemmer()
# word_stemmer = nostemming()


## tokenization
## Im gonna do all three together
X_train_p1 = []
for i in range(len(train['text'])):
    lin = text_to_word_sequence(train['text'][i])
    lin2 = []
    for a in lin:
        if a not in stopwords:
            lin2.append(word_stemmer.stem(lemmatizer.lemmatize(a)))
    if len(lin2) == 0:
        lin2.append("Drakyrus")
    X_train_p1.append(lin2)
    
X_test_p1 = []
for i in range(len(test['text'])):
    lin = text_to_word_sequence(test['text'][i])
    lin2 = []
    for a in lin:
        if a not in stopwords:
            lin2.append(word_stemmer.stem(lemmatizer.lemmatize(a)))
    if len(lin2) == 0:
        lin2.append("Drakyrus")
    X_test_p1.append(lin2)

if GiveOutputFile:
    X_realtest_p1 = []
    for i in range(len(realtest['text'])):
        lin = text_to_word_sequence(realtest['text'][i])
        lin2 = []
        for a in lin:
            if a not in stopwords:
                lin2.append(word_stemmer.stem(lemmatizer.lemmatize(a)))
        if len(lin2) == 0:
            lin2.append("Drakyrus")
        X_realtest_p1.append(lin2)

# print(X_train_p1)
# print(X_test_p1)
```

# Basic Models

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import f1_score

l_train = []
l_test = []
l_realtest=[]
liword = []
for a in X_train_p1:
    b = ' '.join(a)
    if type(b) != type("lsi"):
        print("ghotala")
        print(b)
    l_train.append(b)
for a in X_test_p1:
    b = ' '.join(a)
    if type(b) != type("lsi"):
        print("ghotala")
        print(b)
    l_test.append(b)
    
if GiveOutputFile:
    for a in X_realtest_p1:
        b = ' '.join(a)
        if type(b) != type("lsi"):
            print("ghotala")
            print(b)
        l_realtest.append(b)
    liword = l_train + l_test + l_realtest
else:
    liword = l_train + l_test

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = "word", max_features = 5000)
wordtrain = vectorizer.fit_transform(liword)
wordtrain = wordtrain.toarray()
train_vectorized = vectorizer.transform(l_train)
test_vectorized = vectorizer.transform(l_test)
if GiveOutputFile:
    realtest_vectorized = vectorizer.transform(l_realtest)


```

## Logistic Regression

```python
###### Logistic Regression ########

lr = LogisticRegression()
lr.fit(train_vectorized,train['label_num'])
print(classification_report( lr.predict(test_vectorized) , test['label_num']))
print(accuracy_score( lr.predict(test_vectorized) , test['label_num']))

###### Confusion Matrix ##########
cm = confusion_matrix(lr.predict(test_vectorized) , test['label_num'])
sum = np.sum(cm)
accuracy = (cm[0][0] + cm[1][1] + cm[2][2])/sum
print('-------------Confusion Matrix-----------\n', cm,'\n----------------------------------------')
print('Test accuracy   ',accuracy)

print('-------------    F1 Score   -----------')
score = f1_score(test['label_num'], lr.predict(test_vectorized), average=None)
for i in range(3):
    print(i,round(score[i],2))
print("Average F1 score",round(np.sum(score)/3,3))
```

## Random Forest Method

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(train_vectorized,train['label_num'])

###### Confusion Matrix ##########
cm = confusion_matrix(forest.predict(test_vectorized) , test['label_num'])
sum = np.sum(cm)
accuracy = (cm[0][0] + cm[1][1] + cm[2][2])/sum
print('-------------Confusion Matrix-----------\n', cm,'\n----------------------------------------')
print('Test accuracy   ',accuracy)

print('-------------    F1 Score   -----------')
score = f1_score(test['label_num'], forest.predict(test_vectorized), average=None)
for i in range(3):
    print(i,round(score[i],2))
print("Average F1 score",round(np.sum(score)/3,3))
```

## SVM Model

```python

###### SVM ########

svm = LinearSVC()
svm.fit(train_vectorized,train['label_num'])
print(classification_report( svm.predict(test_vectorized) , test['label_num']))
print(accuracy_score( svm.predict(test_vectorized) , test['label_num']))

###### Confusion Matrix ##########
cm = confusion_matrix(svm.predict(test_vectorized) , test['label_num'])
sum = np.sum(cm)
accuracy = (cm[0][0] + cm[1][1] + cm[2][2])/sum
print('-------------Confusion Matrix-----------\n', cm,'\n----------------------------------------')
print('Test accuracy   ',accuracy)

print('-------------    F1 Score   -----------')
score = f1_score(test['label_num'], svm.predict(test_vectorized), average=None)
for i in range(3):
    print(i,round(score[i],2))
print("Average F1 score",round(np.sum(score)/3,3))
```

# Voting Classifier

```python
## Voting Classifier ( SVM+LR)
estimators = [ ('svm',svm) , ('lr', lr) ]
clf = VotingClassifier(estimators , voting='hard')
clf.fit(train_vectorized,train['label_num'])
print(classification_report( clf.predict(test_vectorized) , test['label_num']))
print(accuracy_score( clf.predict(test_vectorized) , test['label_num']))

###### Confusion Matrix ##########
cm = confusion_matrix(clf.predict(test_vectorized) , test['label_num'])
sum = np.sum(cm)
accuracy = (cm[0][0] + cm[1][1] + cm[2][2])/sum
print('-------------Confusion Matrix-----------\n', cm,'\n----------------------------------------')
print('Test accuracy   ',accuracy)

print('-------------    F1 Score   -----------')
score = f1_score(test['label_num'], clf.predict(test_vectorized), average=None)
for i in range(3):
    print(i,round(score[i],2))
print("Average F1 score",round(np.sum(score)/3,3))
```

## Complex model

```python
Y_train = to_categorical(train['label_num'],num_classes=3)
Y_test = to_categorical(test['label_num'],num_classes=3)
```

```python
##############################################
########## Text part of X_train ##############
##############################################
tokenizer = Tokenizer(num_words=max_features,oov_token='<OOV>')
full_text = list(X_train_p1) + list(X_test_p1)
tokenizer.fit_on_texts(full_text)
print( tokenizer.word_index)
X_train_p1 = pad_sequences( tokenizer.texts_to_sequences(train['text']) , maxlen=max_words)
X_test_p1  = pad_sequences( tokenizer.texts_to_sequences(test['text'])  , maxlen=max_words)

X_train_p1 = pad_sequences(X_train_p1, maxlen=max_words)
X_test_p1 = pad_sequences(X_test_p1, maxlen=max_words)

if GiveOutputFile:
    X_realtest_p1 = pad_sequences(tokenizer.texts_to_sequences(realtest['text']),maxlen=max_words)
    X_realtest_p1 = pad_sequences(X_realtest_p1, maxlen=max_words)
```

# Base model for Text Processing

## GRU model

```python
model1 = Sequential()
model1.add(Embedding(max_features,100,input_length=max_words))
model1.add(GRU(128,return_sequences=True))
model1.add(Dropout(0.4))
model1.add(Flatten())
model1.add(Dense(128,activation='relu'))
model1.add(Dropout(0.4))
model1.add(Dense(32,activation='relu', kernel_regularizer=l2(0.01), bias_regularizer=l2(0.01)))
model1.add(Dropout(0.2))
model1.add(Dense(3,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer=Adam(lr=0.0001),metrics=['accuracy'])
model1.summary()
```

```python
%%time
epoch = 34
batch_size = 128 # changed this
history1=model1.fit(X_train_p1, Y_train, validation_data=(X_test_p1, Y_test),epochs=epoch, batch_size=batch_size, verbose=1)
```
